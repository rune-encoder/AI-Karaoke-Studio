```python
import pickle
from pprint import pprint

# Load the pkl file
with open('aligned_lyrics.pkl', 'rb') as f:
    data = pickle.load(f)

# Print the data
pprint(data,sort_dicts=False)
```

```python
print(len(data))
```

```python
last_end = 0.0  # Initialize the last end time

# chunks = {
#     1: {'start': 0.0, 'end': 63.54},
#     2: {'start': 63.54, 'end': 94.32},
#     3: {'start': 101.26, 'end': 135.42},
#     4: {'start': 135.42, 'end': 166.68},
#     5: {'start': 167.16, 'end': 211.14},
#     6: {'start': 211.14, 'end': 260.26},
#     7: {'start': 260.9, 'end': 261.56},
# }

# Function to find the chunk number for a given time
def find_chunk_number(time, chunks):
    for chunk_number, chunk in chunks.items():
        if chunk['start'] <= time < chunk['end']:
            return chunk_number
    return None


for word in data:
    # chunk_number = find_chunk_number(word.start, chunks)

    if word.start < last_end:
        print(f"❌: {word.word} | {word.start} => {word.end}")
    else:
        print(f"✅: {word.word} | {word.start} => {word.end}")

    if word.end > last_end:
        last_end = word.end

    # if word.end == chunks[chunk_number]['end']:
    #     print(f"=============== End of Chunk {chunk_number} ===============")
```

```python
# Dictionary to collect words for each verse
verses_dict = {}

# Iterate over each word and group them by verse_number
for word_info in data:
    verse_number = word_info.verse_number
    word = word_info.word

    if verse_number not in verses_dict:
        verses_dict[verse_number] = []

    verses_dict[verse_number].append(word)

# List to store the joined verses
verses = []

# Join the words for each verse and append to the verses list
for verse_number in sorted(verses_dict.keys()):
    verse_text = ' '.join(verses_dict[verse_number])
    verses.append(verse_text)

# Print the verses
for verse in verses:
    print(verse)
```
